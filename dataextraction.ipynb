```python
import os
import glob
import xml.etree.ElementTree as ET
import pandas as pd

# Path to your dataset root
dataset_path = "Dataset"

# Collect all XML files
xml_files = glob.glob(os.path.join(dataset_path, "**", "*.xml"), recursive=True)

data = []

for file in xml_files:
    try:
        tree = ET.parse(file)
        root = tree.getroot()

        # Example: Extract <Question> and <Answer> tags
        question = root.findtext(".//Question")  # adjust XPath if needed
        answer = root.findtext(".//Answer")      # adjust XPath if needed

        if question and answer:
            data.append({"file": os.path.basename(file), "question": question, "answer": answer})

    except Exception as e:
        print(f"Error parsing {file}: {e}")

# Convert to DataFrame
df = pd.DataFrame(data)

# Save as CSV
df.to_csv("dataset_QA.csv", index=False, encoding="utf-8")

print(f"Saved {len(df)} records to dataset_QA.csv")
```
